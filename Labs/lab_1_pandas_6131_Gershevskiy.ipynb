```python
data.info()
```

```python
import numpy as np
```

```python
##генерация пропущенных значений
data_with_nan_0 = data.copy()

rand_arr = []
for i in range(100):
    rand_int = np.random.randint(0, data.shape[0])
    rand_arr.append(rand_int)
data_with_nan_0['Course Name'][rand_arr] = np.nan
```

```python
data_with_nan_0.info()
```

```python
## удаление строк с пропусками
data_with_nan_0.dropna(inplace=True)
data_with_nan_0.info()
```

```python
## заполнение пропущенных значений средними показателями

##генерация пропущенных значений
data_with_nan_1 = data.copy()

rand_arr = []
for i in range(100):
    rand_int = np.random.randint(0, data.shape[0])
    rand_arr.append(rand_int)
data_with_nan_1['Course Rating'][rand_arr] = np.nan
```

```python
data_with_nan_1.info()
```

```python
## очищаем поле от str-значений
data_with_nan_1 = data_with_nan_1.loc[data_with_nan_1['Course Rating'] != 'Not Calibrated']
```

```python
med = data_with_nan_1['Course Rating'].median()
med
```

```python
#создание маски из булевых значений, где True - NaN
mask = data_with_nan_1['Course Rating'].isna()
```

```python
#к значениям NaN - присваиваем значение среднего значения по колонке
data_with_nan_1[mask] = med
```

```python
data_with_nan_1.info() #пустных значений нет
```
